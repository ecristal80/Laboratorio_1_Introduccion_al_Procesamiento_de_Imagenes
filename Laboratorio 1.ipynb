PROBLEMA 1
Desarrolle una función la cual reciba dos parámetros, una imagen y un entero llamado color, la función debe devolver una imagen la cual tenga activos los canales de color según los siguientes puntos:
- Si el parámetro color vale 1, la imagen debe mostrar activos únicamente el color azul.
- Si el parámetro color vale 2, la imagen debe mostrar activos únicamente el color verde.
- Si el parámetro color vale 3, la imagen debe mostrar activos únicamente el color rojo.
- Si el parámetro color vale 10, la imagen debe mostrar activos únicamente los colores rojo y verde.
- Si el parámetro color vale 20, la imagen debe mostrar activos únicamente los colores verde y azul.
- Si el parámetro color vale 30, la imagen debe mostrar activos únicamente los colores azul y rojo.

Cargando librerias

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

pip install matplotlib

```python
def seleccionar_color(image_path, color):
    
    # Se carga la imagen
    imagen = cv2.imread(image_path)

    # Se crea una matriz de ceros del mismo tamaño que la imagen
    resultado = np.zeros_like(imagen)

    if color == 1:  # Mostrar solo el color azul
        resultado[:, :, 0] = imagen[:, :, 0]
    elif color == 2:  # Mostrar solo el color verde
        resultado[:, :, 1] = imagen[:, :, 1]
    elif color == 3:  # Mostrar solo el color rojo
        resultado[:, :, 2] = imagen[:, :, 2]
    elif color == 10:  # Mostrar rojo y verde
        resultado[:, :, 0] = imagen[:, :, 0]
        resultado[:, :, 1] = imagen[:, :, 1]
    elif color == 20:  # Mostrar verde y azul
        resultado[:, :, 1] = imagen[:, :, 1]
        resultado[:, :, 2] = imagen[:, :, 2]
    elif color == 30:  # Mostrar azul y rojo
        resultado[:, :, 0] = imagen[:, :, 0]
        resultado[:, :, 2] = imagen[:, :, 2]
    else:
        raise ValueError("Valor de color no válido")

    return resultado
```

PROBLEMA 2
En el .zip del laboratorio se le compartió un conjunto de imágenes en escala de grises, estas imágenes fueron creadas utilizando una escala de grises en 3D, cree una función que dadas las 3 imágenes se construya la imagen original a color.

```python
def construir_imagen_color(imagen_roja, imagen_verde, imagen_azul):
    # Cargar las tres imágenes en escala de grises
    img_r = cv2.imread(imagen_roja, cv2.IMREAD_GRAYSCALE)
    img_g = cv2.imread(imagen_verde, cv2.IMREAD_GRAYSCALE)
    img_b = cv2.imread(imagen_azul, cv2.IMREAD_GRAYSCALE)

    # Crear una imagen en color a partir de las tres imágenes en escala de grises
    imagen_color = np.zeros((img_r.shape[0], img_r.shape[1], 3), dtype=np.uint8)
    imagen_color[:, :, 0] = img_b  # Canal azul
    imagen_color[:, :, 1] = img_g  # Canal verde
    imagen_color[:, :, 2] = img_r  # Canal rojo

    return imagen_color
```

Ejemplo de construción de imagen

```python
imagen_original = construir_imagen_color("imagen1_salida_gray_rojo.jpg", "imagen1_salida_gray_verde.jpg", "imagen1_salida_gray_azul.jpg")

# Mostrar imagen
plt.imshow(cv2.cvtColor(imagen_original, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Ocultar ejes
plt.show()
```

PROBLEMA 3
Cree una función que dada una imagen cree una escala de grises en tres dimensiones, tome en cuenta que su función debe crear 3 imágenes como salida. Para entregar este ejercicio debe incluir una las imágenes que haya utilizado como prueba y el resultado de las misma, no puede utilizar la imagen del Problema #2.

```python
def crear_escala_de_grises_3D(image_path):
    # Cargar la imagen en color
    imagen_original = cv2.imread(image_path)

    # Separar los canales de color
    canal_rojo = imagen_original[:, :, 2]  # Canal rojo
    canal_verde = imagen_original[:, :, 1]  # Canal verde
    canal_azul = imagen_original[:, :, 0]   # Canal azul

    # Crear imágenes en escala de grises en tres dimensiones
    imagen_gris_rojo = cv2.cvtColor(canal_rojo, cv2.COLOR_GRAY2BGR)
    imagen_gris_verde = cv2.cvtColor(canal_verde, cv2.COLOR_GRAY2BGR)
    imagen_gris_azul = cv2.cvtColor(canal_azul, cv2.COLOR_GRAY2BGR)

    return imagen_gris_rojo, imagen_gris_verde, imagen_gris_azul
```

PROBLEMA 4
Cree una función que dada una imagen, muestre el histograma de cada canal de color y el de escala de grises (utilice un promedio aritmético para su escala de grises, no puede usar funciones de opencv), sus histogramas deben incluir una línea vertical la cual muestre el valor de la media de la distribución.

```python
def mostrar_histogramas(image_path):
    # Cargar la imagen en color
    imagen_original = cv2.imread(image_path)

    # Convertir la imagen en escala de grises utilizando promedio aritmético
    imagen_gris = np.mean(imagen_original, axis=2, dtype=np.uint8)

    # Calculo de histograma
    hist_rojo = cv2.calcHist([imagen_original], [2], None, [256], [0, 256])
    hist_verde = cv2.calcHist([imagen_original], [1], None, [256], [0, 256])
    hist_azul = cv2.calcHist([imagen_original], [0], None, [256], [0, 256])
    hist_gris = cv2.calcHist([imagen_gris], [0], None, [256], [0, 256])

    # Calculo de media de la distribución en escala de grises
    media_gris = np.mean(hist_gris)

    # Crear un arreglo de valores para el eje X (0-255)
    valores_x = np.arange(256)

    # Representar Histogramas
    plt.figure(figsize=(12, 6))

    plt.subplot(221)
    plt.plot(valores_x, hist_rojo, color='red')
    plt.axvline(x=media_gris, color='blue', linestyle='--')
    plt.title("Histograma Canal Rojo")

    plt.subplot(222)
    plt.plot(valores_x, hist_verde, color='green')
    plt.axvline(x=media_gris, color='blue', linestyle='--')
    plt.title("Histograma Canal Verde")

    plt.subplot(223)
    plt.plot(valores_x, hist_azul, color='blue')
    plt.axvline(x=media_gris, color='blue', linestyle='--')
    plt.title("Histograma Canal Azul")

    plt.subplot(224)
    plt.plot(valores_x, hist_gris, color='gray')
    plt.axvline(x=media_gris, color='blue', linestyle='--')
    plt.title("Histograma Escala de Grises")

    plt.tight_layout()
    plt.show()
```

```python
mostrar_histogramas("imagen_prob1.jpg")
```
